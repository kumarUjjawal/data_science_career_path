## Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
axs = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x), color='green',linestyle='--');
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x), color='green',linestyle='--');
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.sin(x))
plt.title("Sin Curve")
plt.xlabel('x')
plt.ylabel('Sin(x)')
```
