## Implementations of Commonly Used Activation Functions In Pytorch

### Sigmoid

f(x) = 1/1+e^-x

```python
%matplotlib inline
import torch
import matplotlib.pyplot as plt

x = torch.arange(-5.,5., 0.1)
y = torch.sigmoid(x)
plt.plot(x.numpy(), y.numpy());
```

### Tanh
f(x) = tanh x = e^x - e^-x / e^x + e^-x

Values in the range: [-1 , +1]

```python
y = torch.tanh(x)
plt.plot(x.numpy(), y.numpy());
```

### ReLU
f(x) = max(0,x)

So, all a ReLU unit is doing is clipping the negative values to zero

```python
relu = torch.nn.ReLU()

y = relu(x)
plt.plot(x.numpy(), y.numpy());
```

### Softmax
![alt text](softmax.png "Softmax")
Softmax operation divides each output by the sum of all the outputs, which gives us a discrete probability distribution3 over k possible classes.

```python
softmax = torch.nn.Softmax(dim=1)
x_ = torch.rand(1,3)
y_ = softmax(x_)
print(x_)
print(y_)
print(torch.sum(y_, dim=1))
```
